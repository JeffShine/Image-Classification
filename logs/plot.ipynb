```python
import pandas as pd
filename = 'bs1024_lr0.001_optAdam'
# CSV
df = pd.read_csv(filename+'.csv', header=None, names=['Index', 'Value1', 'Error1', 'Value2', 'Error2'])

# DataFrame
# print(df)
import matplotlib.pyplot as plt

# 
index = df['Index']
train_loss = df['Value1']
train_acc = df['Error1']
test_loss = df['Value2']
test_acc = df['Error2']

# 
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# train-loss  test-loss
axs[0].plot(index, train_loss, '-o', label='train')
axs[0].plot(index, test_loss, '-o', label='test')
axs[0].set_title('Loss')
axs[0].legend()

# train-acc  test-acc
axs[1].plot(index, train_acc, '-o', label='train')
axs[1].plot(index, test_acc, '-o', label='test')
axs[1].set_title('Accuracy')
axs[1].legend()
# 
filename = filename.replace('bs', 'batch-size=').replace('lr', 'lr=').replace('opt', 'optimizer=').replace('_', ', ')
fig.text(0.5, -0.03, filename, ha='center', fontsize=12)
# 
plt.tight_layout()
plt.show()
```
