```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Define the directory containing the CSV files
logs_directory = './'

# List all CSV files in the directory
csv_files = [f for f in os.listdir(logs_directory) if f.endswith('.csv')]

for filename in csv_files:
    # Read CSV file
    df = pd.read_csv(os.path.join(logs_directory, filename), header=None, names=['Index', 'Value1', 'Error1', 'Value2', 'Error2'])
    
    # Read each column
    index = df['Index']
    train_loss = df['Value1']
    train_acc = df['Error1']
    test_loss = df['Value2']
    test_acc = df['Error2']

    # Create a figure and two subplots
    fig, axs = plt.subplots(1, 2, figsize=(8, 4))

    # Plot the first line chart (train-loss and test-loss)
    axs[0].plot(index, train_loss, '-o', label='train')
    axs[0].plot(index, test_loss, '-o', label='test')
    axs[0].set_title('Loss')
    axs[0].legend()

    # Plot the second line chart (train-acc and test-acc)
    axs[1].plot(index, train_acc, '-o', label='train')
    axs[1].plot(index, test_acc, '-o', label='test')
    axs[1].set_title('Accuracy')
    axs[1].legend()

    # Set the main title
    plot_title = filename.replace('bs', 'batch-size=').replace('lr', 'lr=').replace('opt', 'optimizer=').replace('.csv', '').replace('_', ', ')
    fig.suptitle(plot_title)

    # Adjust layout
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)  # Adjust top to make room for the title

    # Show the plot
    plt.show()
```
